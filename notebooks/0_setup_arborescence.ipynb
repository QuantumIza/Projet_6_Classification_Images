# ⚙️ Notebook 0 – Initialisation du projet

Ce notebook crée l’arborescence complète du projet `Projet_6_Classification_Images` dans Google Drive.

Il monte le Drive, crée les dossiers nécessaires, et initialise les fichiers `README.md` et `requirements.txt`.

À exécuter une seule fois au début du projet, ou en cas de réinitialisation.


```python
# ---------------------------
# --- MONTER LE GOOGLE DRIVE
# ---------------------------
from google.colab import drive
drive.mount('/content/drive')
```

```python
# --- CREATION ARBORESCENCE PROJET
import os

# Chemin racine du projet dans ton Drive
base_path = '/content/drive/MyDrive/Projet_6_Classification_Images'

# Sous-dossiers à créer
folders = [
    'data/raw',
    'data/processed',
    'notebooks',
    'src',
    'outputs/models',
    'outputs/logs',
    'outputs/predictions',
    'streamlit_app'
]

# Création des dossiers
for folder in folders:
    path = os.path.join(base_path, folder)
    os.makedirs(path, exist_ok=True)

# Fichiers vides à créer
files = [
    'README.md',
    'requirements.txt'
]

for file in files:
    file_path = os.path.join(base_path, file)
    with open(file_path, 'w') as f:
        f.write(f'# {file} – à compléter\n')

print("# --- ARBORESCENCE DU PROJET CREEE AVEC SUCCES !")
```
